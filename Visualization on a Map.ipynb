```python
import pandas as pd
from shapely.geometry import Point
import geopandas as gpd
from geopandas import GeoDataFrame
```

```python
from datetime import date
from datetime import datetime
```

```python
df = pd.read_csv('Longitude_Latitude.csv')

gmtr = [Point(xy) for xy in zip(df['Longitude'],df['Latitude'])]
```

```python
gdf = GeoDataFrame(df, geometry=gmtr) 
```

```python
gdfMilitary = gdf[gdf['Operator'].str.contains('Military', na=False)]
gdfNonMilitary = gdf.drop(gdfMilitary.index)
#data = data.drop(data[data['TotalNumberOfDead'] < 710].index)
```
