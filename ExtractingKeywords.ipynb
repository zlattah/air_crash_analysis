```python
import sklearn
import pandas as pd
import numpy as np
import graphviz
from string import punctuation
import spacy

nlp = spacy.load("en_core_web_sm")

data = pd.read_csv("Airplane_Crashes_and_Fatalities_Since_1908.csv")
data = pd.DataFrame(data)

summaries = data.pop('Summary')
```

```python
def get_hotwords(text):
    res = []
    wordfreq = {}
    pos_tag = ['PROPN', 'ADJ', 'NOUN']
    doc = ""
    if type(text) == type("abc"):
        doc = nlp(text.lower())
    for token in doc:
        if(token.text in nlp.Defaults.stop_words or token.text in punctuation):
            continue
        if(token.pos_ in pos_tag):
            res.append(token.text)
    
    for word in res:
        if word not in wordfreq:
            wordfreq[word] = 0
        wordfreq[word]+=1
    
    return wordfreq
```

```python
all_keywords = set()
frequency = []

for txt in summaries:
    frequency.append(get_hotwords(txt))
    output = frequency[-1].keys()
    all_keywords = all_keywords.union(set(output))
        

#print(all_keywords)
```

```python
zeros = [0]*(summaries.size)
new_data = {}
for keyword in all_keywords:
    new_data[keyword] = zeros.copy()

```

```python
for i in range(summaries.size):
    for word in frequency[i].keys():
        new_data[word][i] = frequency[i][word]
```

```python
datfr = pd.DataFrame.from_dict(new_data)
datfr
```

```python
datacopy = data.copy()
```

```python
newdatacopy = datfr.copy()
```

```python
final = pd.merge(datacopy, newdatacopy, left_index = True, right_index = True)
```
