```python
import pandas as pd
from geopy.geocoders import Nominatim

loc = Nominatim(user_agent = 'GetLoc')

data = pd.read_csv("Airplane_Crashes_and_Fatalities_Since_1908.csv")
data = pd.DataFrame(data)
```

```python
newdata = {'Latitude':[], 'Longitude':[]}

for location in data['Location']:
    try:
        getLoc = loc.geocode(location)
        if getLoc != None:
            newdata['Latitude'].append(getLoc.latitude)
            newdata['Longitude'].append(getLoc.longitude)
        else:
            newdata['Latitude'].append(-1)
            newdata['Longitude'].append(-1)
    except:
        newdata['Latitude'].append(-1)
        newdata['Longitude'].append(-1)
```
